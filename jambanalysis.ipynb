```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example data

df = pd.read_csv('jam.csv')

# Boxplot
sns.boxplot(x='Gender', y='JAMB_Score', data=df)
plt.title("Boxplot of Exam Scores by Gender")
plt.show()

# Calculate IQR and identify outliers
Q1 = df.groupby('Gender')['JAMB_Score'].quantile(0.25)
Q3 = df.groupby('Gender')['JAMB_Score'].quantile(0.75)
IQR = Q3 - Q1

# Detect outliers
outliers = df[
    (df['JAMB_Score'] < (Q1[df['Gender']].values - 1.5 * IQR[df['Gender']].values)) |
    (df['JAMB_Score'] > (Q3[df['Gender']].values + 1.5 * IQR[df['Gender']].values))
]

print("Outliers:")
print(outliers)
```

```python
import pandas
from sklearn import linear_model

df = pandas.read_csv("jam.csv")

Xaxis = df[['Study_Hours_Per_Week', 'Attendance_Rate','Teacher_Quality','Distance_To_School']]


y = df['JAMB_Score']

regr = linear_model.LinearRegression()
regr.fit(Xaxis, y)
print(regr.fit(Xaxis, y))
#predict the student performance in jamb:
predictedscores = regr.predict([[30, 1300,4,11]])

print(predictedscores)
```
